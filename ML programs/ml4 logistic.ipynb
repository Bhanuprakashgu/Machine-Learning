```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
da=pd.read_csv('HR_comma_sep.csv')
```

```python
da
```

```python
dum=pd.get_dummies(da.salary)
```

```python
dum
```

```python
data=pd.concat([da,dum],axis='columns')
```

```python
data
```

```python
x=data[['satisfaction_level','average_montly_hours','promotion_last_5years','high','low','medium']]
```

```python
x
```

```python
y=data.left
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
