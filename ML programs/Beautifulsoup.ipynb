```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url="https://www.w3schools.com/python/python_syntax.asp"
html=urlopen(url)
```

```python
soup=BeautifulSoup(html)
```

```python
title=soup.title
print(title)
print(title.text)
```

```python
links=soup.find_all('a',href=True)
for link in links:
  print(link['href'])
```
